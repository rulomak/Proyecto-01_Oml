## Quito columnas para optimizar recursos en el deploy

```python
import pandas as pd
```

```python
# cargo el DataSet
df_movies = pd.read_csv('movies_etl_api.csv')
```

```python
pd.options.display.max_columns = None   # para ver todas las columnas 
```

```python
df_movies.head()
```

- columnas que no se utilizan:   
id , original_language, popularity, release_date, 
runtime, status, tagline, vote_average, name_languages

```python
columnas_innecesarias = ['id' , 'original_language', 'popularity', 'release_date', 'runtime', 'status', 'tagline', 'vote_average', 'name_languages']

df_movies = df_movies.drop(columnas_innecesarias, axis=1)
```

```python
# cambio la data que se exporto nula a sin datos 
df_movies['name_production_compani'] = df_movies['name_production_compani'].fillna('sin datos')
df_movies['country_name'] = df_movies['country_name'].fillna('sin datos')
```

```python
df_movies.info()
```

```python
df_movies.head()
```

## Exporto nuevamente a csv 

```python
df_movies.to_csv('movies_etl_api.csv', index=False)
```
